```python
def depth_limited_dfs(maze, start, goal, limit):
    rows = len(maze)
    cols = len(maze[0])

    stack = [(start, [start], 0)]  # (current_position, path, depth)
    visited = []

    while stack:
        (x, y), path, depth = stack.pop()

        if (x, y) == goal:
            return path

        if depth < limit:
            visited.append((x, y))

            # Possible moves: Up, Down, Left, Right
            moves = [(x-1,y), (x+1,y), (x,y-1), (x,y+1)]

            for nx, ny in moves:
                if (0 <= nx < rows and 0 <= ny < cols and
                    maze[nx][ny] != 1 and
                    (nx, ny) not in visited):

                    stack.append(((nx, ny), path + [(nx, ny)], depth + 1))

    return None
```

```python
def iterative_deepening_search(maze, start, goal):
    depth = 0

    while True:
        result = depth_limited_dfs(maze, start, goal, depth)
        if result is not None:
            return result
        depth += 1
```

```python
maze = [
['S', 0, 0, 1, 0],
[1, 0, 1, 0, 0],
[0, 0, 0, 0, 'G'],
[1, 1, 0, 1, 1]
]

# Locate Start and Goal
for i in range(len(maze)):
    for j in range(len(maze[0])):
        if maze[i][j] == 'S':
            start = (i, j)
        if maze[i][j] == 'G':
            goal = (i, j)

# Run IDS
path = iterative_deepening_search(maze, start, goal)

print("Path to Goal:", path)
```

# Behavior of Depth-First Search (DFS):

DFS explores one path deeply before backtracking.

* If it chooses a wrong branch:

* It may go very deep into a dead end.

* If depth limit is too small → goal not found.

* If limit is too large → unnecessary exploration.

DFS is:

* Memory efficient

* Not optimal

* Not guaranteed to find shallowest goal
---
Behavior of Iterative Deepening Search (IDS)

- Repeats DFS with increasing depth limits.

- Explores shallow nodes first.

- Avoids getting stuck in deep dead ends.

- Requires memory similar to DFS.

- Guarantees finding the goal if it exists.

IDS is better because:

1. It does not assume goal depth.

2. It balances memory efficiency and completeness.

3. It systematically expands the search frontier.

For an unknown maze with no heuristic information and possible deep dead ends, Iterative Deepening Search is more suitable than simple DFS because it guarantees finding the goal while maintaining low memory usage and avoiding excessive deep exploration.
