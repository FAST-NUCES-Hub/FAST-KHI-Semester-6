```python
import random

class Env:
    def __init__(self):
        self.c = list("ABCDEFGHI")
        self.s = {}
        for i in self.c:
            self.s[i] = random.choice(["safe", "vul"])

    def show(self):
        for i in self.c:
            print("Component", i, "is", self.s[i])


class Agent:
    def __init__(self, env):
        self.e = env
        self.v = []

    def scan(self):
        print("\nSYSTEM SCAN\n")
        for i in self.e.c:
            if self.e.s[i] == "vul":
                print("Alert:", i, "is vulnerable")
                self.v.append(i)
            else:
                print("OK:", i, "is secure")

    def patch(self):
        print("\nPATCHING\n")
        for i in self.v:
            self.e.s[i] = "safe"
            print("Fixed:", i, "is now secure")


def agent_run():
    e = Env()
    print("INITIAL SYSTEM CHECK\n")
    e.show()

    a = Agent(e)
    a.scan()
    a.patch()

    print("\nFINAL STATUS\n")
    e.show()
```

```python
agent_run()
```

```python
class Server:
    def __init__(self, name, load):
        self.name = name
        self.load = load

class LoadBalancerAgent:
    def balance(self, servers):
        overloaded = []
        underloaded = []

        for s in servers:
            if s.load == "Overloaded":
                overloaded.append(s)
            elif s.load == "Underloaded":
                underloaded.append(s)

        for o in overloaded:
            if underloaded:
                u = underloaded.pop(0)
                o.load = "Balanced"
                u.load = "Balanced"
```

```python
servers = [
    Server("Server1", "Overloaded"),
    Server("Server2", "Underloaded"),
    Server("Server3", "Balanced"),
    Server("Server4", "Overloaded"),
    Server("Server5", "Underloaded")
]

agent = LoadBalancerAgent()
agent.balance(servers)

for s in servers:
    print(s.name, "-", s.load)
```

```python
class BackupAgent:
    def retry_failed(self, tasks):
        for i in range(len(tasks)):
            if tasks[i] == "Failed":
                tasks[i] = "Completed"
```

```python
tasks = ["Completed", "Failed", "Completed", "Failed", "Completed"]

agent = BackupAgent()
agent.retry_failed(tasks)

for t in tasks:
    print(t)
```

```python
import random

class SecurityAgent:
    def __init__(self, system):
        self.system = system

    def scan_and_patch(self):
        for comp in self.system:
            if self.system[comp] == "Low":
                print(comp, "Low Risk Vulnerability detected and patched")
                self.system[comp] = "Safe"
            elif self.system[comp] == "High":
                print(comp, "High Risk Vulnerability - Premium Service Required")
            else:
                print(comp, "is Safe")
```

```python
components = ["A","B","C","D","E","F","G","H","I"]
states = ["Safe","Low","High"]

system = {}
for c in components:
    system[c] = random.choice(states)

print("Initial State:")
for k in system:
    print(k, "-", system[k])

agent = SecurityAgent(system)
print("\nScanning System:")
agent.scan_and_patch()

print("\nFinal State:")
for k in system:
    print(k, "-", system[k])
```

```python
class DeliveryRobot:
    def __init__(self):
        self.location = "Storage"

    def move(self, place):
        self.location = place
        print("Moved to", place)

    def pick_medicine(self, medicine):
        print("Picked up", medicine)

    def scan_id(self, patient):
        print("Scanned ID of", patient)

    def deliver(self, patient):
        print("Delivered medicine to", patient)

    def alert_staff(self):
        print("Alerted medical staff")
```

```python
robot = DeliveryRobot()

robot.move("Medicine Storage")
robot.pick_medicine("Paracetamol")

robot.move("Room 101")
robot.scan_id("Patient 101")
robot.deliver("Patient 101")

robot.move("Nurse Station")
robot.alert_staff()
```

```python
rooms = {
    "a":"safe","b":"safe","c":"fire",
    "d":"safe","e":"fire","f":"safe",
    "g":"safe","h":"safe","j":"fire"
}

path = ["a","b","c","d","e","f","g","h","j"]

def display():
    grid = ["a","b","c","d","e","f","g","h","j"]
    for i in range(0,9,3):
        row = ""
        for r in grid[i:i+3]:
            if rooms[r] == "fire":
                row += "F "
            else:
                row += ". "
        print(row)
    print()
```

```python
for room in path:
    print("Robot moved to", room)
    if rooms[room] == "fire":
        print("Fire detected! Extinguishing...")
        rooms[room] = "safe"
    else:
        print("Room is safe")
    display()

print("Final Environment Status:")
display()
```
