Q1:

```python
class SecurityAgent:
    def __init__(self, agent_id, name, status):
        self.agent_id = agent_id
        self.name = name
        self.status = status

class FirewallAgent(SecurityAgent):
    def monitor_traffic(self):
        print(self.name, "is monitoring network traffic")

class MalwareDetectionAgent(SecurityAgent):
    def scan_files(self):
        print(self.name, "is scanning files for malware")

class AutomationAgent(SecurityAgent):
    def run_automation(self):
        print(self.name, "is running security automation tasks")
```

```python
f = FirewallAgent(1, "FirewallAgent1", "Active")
m = MalwareDetectionAgent(2, "MalwareAgent1", "Active")
a = AutomationAgent(3, "AutomationAgent1", "Active")

f.monitor_traffic()
m.scan_files()
a.run_automation()
```

```python
class Threat:
    def __init__(self, threat_id, name, severity):
        self.threat_id = threat_id
        self.name = name
        self.severity = severity

class PhishingThreat(Threat):
    def analyze_email(self):
        print(self.name, "is analyzing suspicious emails")

class RansomwareThreat(Threat):
    def scan_files(self):
        print(self.name, "is scanning files for ransomware")

class BotnetThreat(Threat):
    def detect_traffic(self):
        print(self.name, "is detecting unusual network traffic")
```

```python
p = PhishingThreat(1, "PhishingAttack", "High")
r = RansomwareThreat(2, "RansomwareAttack", "Critical")
b = BotnetThreat(3, "BotnetAttack", "Medium")

p.analyze_email()
r.scan_files()
b.detect_traffic()
```

Q3:

```python
class ResponseAgent:
    def execute_response(self):
        print("Executing response")

class AlertAgent(ResponseAgent):
    def execute_response(self):
        print("Sending alert notification")

class BlockAgent(ResponseAgent):
    def execute_response(self):
        print("Blocking malicious activity")

class RecoverAgent(ResponseAgent):
    def execute_response(self):
        print("Recovering affected system")
```

```python
agents = [AlertAgent(), BlockAgent(), RecoverAgent()]

for agent in agents:
    agent.execute_response()
```

Q4:

```python
class Employee:
    def work(self):
        print("Employee is working")

class Manager(Employee):
    def work(self):
        print("Manager is managing the team")

class Developer(Employee):
    def work(self):
        print("Developer is writing code")

class Designer(Employee):
    def work(self):
        print("Designer is creating designs")
```

```python
employees = [Manager(), Developer(), Designer()]

for emp in employees:
    emp.work()
```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.__grade = 0

    def set_grade(self, grade):
        self.__grade = grade

    def get_grade(self):
        return self.__grade

    def display_info(self):
        print("Name:", self.name)
        print("Grade:", self.__grade)
```

```python
s1 = Student("Muzammil")

s1.set_grade(67)
print(s1.get_grade())
s1.display_info()
```
